```python
import numpy as np
import cv2
import random
```

```python
img = cv2.imread('C:/Documents-Summer2018/CVIP/PA2/Butterfly.jpg')
row,col,colorsp = img.shape
segmented_img = np.zeros((img.shape))
print(segmented_img.shape)
ref_table = np.zeros((row * col + 1,5))
x,y = ref_table.shape
print(ref_table.shape)
h = 90
iters = 40
```

```python
#rearranging into feature matrix
t = 0
for i in range(0,row):
    for j in range(0,col):
        ref_table[t][0]=img[i][j][0]
        ref_table[t][1]=img[i][j][1]
        ref_table[t][2]=img[i][j][2]
        ref_table[t][3]=i
        ref_table[t][4]=j
        t+=1        
        
    
```

```python
print (x,len(ref_table))
print(img)
print(ref_table)
```

```python
print('Initial length of Reference table :',len(ref_table))
iteration_no = 1

while(len(ref_table)>0):
    print('Current length of Reference table :',len(ref_table))
    print('Iter : ',iteration_no)
    iteration_no+=1    
    fetched = random.randint(0,len(ref_table)-1)#https://www.pythoncentral.io/how-to-generate-a-random-number-in-python/
        
    dist = 0;
    Cluster_ls = []

    for i in range(0,len(ref_table)):
        dist = np.linalg.norm(ref_table[fetched]-ref_table[i])#https://stackoverflow.com/questions/1401712/how-can-the-euclidean-distance-be-calculated-with-numpy
        if dist < h:
            Cluster_ls.append(i)
    print('Cluster Size:',len(Cluster_ls))
    if(len(Cluster_ls) > 0):
        m1,m2,m3,m4,m5 = 0,0,0,0,0
        for i in range(0,len(Cluster_ls)):
            idx = Cluster_ls[i]
            m1 += ref_table[idx][0]
            m2 += ref_table[idx][1]
            m3 += ref_table[idx][2]
            m4 += ref_table[idx][3]
            m5 += ref_table[idx][4]

        
        
        newMean = np.zeros((1,5))
        l1 = len(Cluster_ls)    
        newMean[0][0] = m1/l1
        newMean[0][1] = m2/l1
        newMean[0][2] = m3/l1
        newMean[0][3] = m4/l1
        newMean[0][4] = m5/l1

                  
        new_dist = np.linalg.norm(ref_table[fetched]-newMean[0])#https://stackoverflow.com/questions/1401712/how-can-the-euclidean-distance-be-calculated-with-numpy
  
#         print('New Distance :',new_dist)


        if(new_dist <= iters):
            for i in range(0,len(Cluster_ls)):
                index = Cluster_ls[i]
                x = int(ref_table[index][3])
                y = int(ref_table[index][4])
                segmented_img[x][y][0] = newMean[0][0]
                segmented_img[x][y][1] = newMean[0][1]
                segmented_img[x][y][2] = newMean[0][2]

            ref_table = np.delete(ref_table,Cluster_ls,0)

        else:
            x = int(ref_table[fetched][3])                
            y = int(ref_table[fetched][4])
            segmented_img[x][y][0] = ref_table[fetched][0]
            segmented_img[x][y][1] = ref_table[fetched][1]
            segmented_img[x][y][2] = ref_table[fetched][2]
            ref_table = np.delete(ref_table,fetched,0)           



print(segmented_img)
      
cv2.imshow("Segmented Image:",segmented_img/255)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('C:/Documents-Summer2018/CVIP/PA2/MS/SegmentedImg_90_40.png',segmented_img)
        
```
